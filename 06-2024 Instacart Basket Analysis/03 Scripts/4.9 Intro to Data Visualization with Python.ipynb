# Creating Line Charts

## Sampling Data

```python
# create a list holding True/False values to test the np.random.rant() <= 0.7
np.random.seed(4)
dev = np.random.rand(len(df)) <= 0.7
```

```python
# sotre 70% of the sample in the datagrame big
big = df[dev]
```

```python
# store 30% of the sample in the dataframe small
small = df[~dev]
```

```python
len(df)
```

```python
len(big) + len(small)
```

```python
df_2 = small[['orders_day_of_week','prices']]
```

## Line Chart

```python
line = sns.lineplot(data = df_2, x = 'orders_day_of_week',y = 'prices')
```

```python
line.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'line_ordersDOW_prices.png'))
```
