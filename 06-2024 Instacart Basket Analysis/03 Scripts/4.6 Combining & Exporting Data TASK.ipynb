#### since both dataframes have 'product_id' in common, we can join the 2 based on that key.  But first we need to get rid of the unnecessary columns in both dataframes

```python
# drop unnamed columns in df_prods

df_prods_preptojoin = df_prods.drop(columns = ['Unnamed: 0'])
```

```python
df_prods_preptojoin.head()
```

```python
# drop unnamed columns and _merge column in df_ords

df_ords_preptojoin = df_ords.drop(columns = ['Unnamed: 0.1', 'Unnamed: 0', '_merge'])
```

```python
df_ords_preptojoin.head()
```

```python
# merge df_prods_preptojoin and df_ords_preptojoin

df_merged = df_prods_preptojoin.merge(df_ords_preptojoin, on = 'product_id', indicator = True)
```
